```python
# barplot  is the rectangular plot  it ois a categorical versus numrical 
import numpy  as np
import matplotlib.pyplot as plt 
x=np.array(["A","B","C","D","E"])
y=np.array([34,45,55,67,87])
plt.bar(x,y)
plt.show()
```

```python
#Draw a barplot bike names and cost of it
import matplotlib.pyplot as plt
y=[10,20,30,40]
x=["yamaha","royalenfield","Pulsar","honda"]
plt.bar(x,y,edgecolor="black")
plt.show()
```

```python
# Draw a barplot bike names and cost of it\,
import matplotlib.pyplot as plt
y=[10,20,30,40]
x=["yamaha","royalenfield","Pulsar","honda"]
plt.bar(x,y,edgecolor="black",color="Red")
plt.show()
```

```python
# Draw a barplot bike names and cost of it\n",
import matplotlib.pyplot as plt
y=[10,20,30,40]
x=["yamaha","royalenfield","Pulsar","honda"]
z=["red","green","Hotpink","yellow"]
plt.bar(x,y,edgecolor="black",color=z)
plt.show()
# There are two types of barplot\n",
# Horizontal plot and verticalplot"
   
```

```python
#Draw a barplot bike names and cost of it
#Horizontal plot
import matplotlib.pyplot as plt
y=[10,20,30,40]
x=["yamaha","royalenfield","pulsar","honda"]
plt.barh(x,y,edgecolor="black",color=z)
plt.show()
   
```

```python
#How to add label name
x=["apple","Orange","kiwi","banana","jackfruit"]
y=[100,56,150,78,100]
import matplotlib.pyplot as plt
plt.xlabel("Different fruit names")
plt.ylabel("cost of the fruits in rupees")
plt.title("Barplot of fruit versus cost")
plt.bar(x,y,edgecolor="Black",color=["red","green","black","orange","yellow"])
plt.show()
# assignment for mon:draw a bar plot vegetable and cost(functions)"
   
```

```python
x = ["i10","i20","BMW","kia","TataNexon"]
y = [10, 20, 30, 40, 50]
colors = ["red", "green", "blue", "violet","hotpink"]
plt.ylabel("Different Car Names")
plt.xlabel("Cost of Different Cars in Dollars")
plt.title("Barplot of Car Names vs Cost in Dollars")
plt.bar(x, y, edgecolor="black", color=colors)
ax = plt.gca()
ax.set_facecolor('lightgrey')  # <-- fixed this line
plt.show()
   
```
