```python
# #A piplot also knows as pierchart or circle chart 
# it is avisual represerentation of data shows proportion of different categoricales
# within a whole
# slices of the pie represent a category and size of the slice is proportional to the 
# percentage of category in  overall dataset
#categorical versus numerical
```

```python
import matplotlib .pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
#lables
import matplotlib .pyplot as plt
import numpy as np
x=["apply","banana","cherry","kiwi"]
y=np.array([35,25,25,15])
plt.pie(y,labels=x,autopct="%2.1f%%")
plt.show()
#autopct is an and precentage to your pie wedge
#labels is a inbuild function to pass categorical inside piechart
```

```python
import matplotlib .pyplot as plt
#import numpy as np
y=["beans","carrot","cabbage","onion","potato"]
x=([100,150,200,200,100])
z=[0,0,0.2,0,0]
W=["Green","hotpink","yellow","red","orange"]
plt.title("Vegitable cost pichart",fontsize=23)
plt.pie(x,labels=y,autopct="%2.1f%%", explode=z,colors=W)
plt.legend(loc="lower left")
plt.show()
#750/x values for get precentage
#categotical always passed by using lables
#always give numrical data  firest and than with catogori cal data   for autoptc 2 is fixed  
```

```python
#how to convert 2D to 3D 
#shadow-3D view
import matplotlib .pyplot as plt
#import numpy as np
y=["beans","carrot","cabbage","onion","potato"]
x=([100,150,200,200,100])
z=[0,0,0.2,0,0]
W=["Green","hotpink","yellow","red","orange"]
plt.title("Vegitable cost pichart",fontsize=23)
plt.pie(x,labels=y,autopct="%2.1f%%", explode=z,colors=W,shadow=True)
plt.legend(loc="lower left")
plt.show()
#750/x values for get precentage
#categotical always passed by using lables
#always give numrical data  firest and than with catogori cal data   for autoptc 2 is fixed  
```

```python
# #Assiggnment :A tearcher serveys how  many are intrested in which games 
# x={"Football","hockey,cricet,valleyball,badmintan}
#    y=[10,5,5,10,10] explode one game, use legent (plot pieplot as well barplot]

import matplotlib.pyplot as h
x=["Football","hockey","cricet","valleyball","badmintan"]
z=["Green","blue","Red","yellow","orange"]
y=[10,5,5,10,15]
w=[0,0,0,0,0.3]
h.pie(y,labels=x,autopct="%2.1f%%",explode=w,shadow=True,colors=z)
h.title("Tearcher Servey")
h.legend(loc="lower left")
h.show()

```

```python
#start angle parameter is difine with an angle in degrees,default angle is 0 zero
#r=red,g-green,b-blue,c-cyan,m-magenta,y=yello,b-black,w-white
import matplotlib .pyplot as plt
y =[100,120,150,200,250,300,]
x=["Rashmali cake","chees cake","chocolate cake","chocolate browwni","vanela browwni","stowbareey with chocolate"]
z=["r","g","b","c","m","y"]
w=[0,0,0,0,0,0.2]
#z=["r"ed,"g"reen,"b"lue,"c"yan,"m"agenta,"y"ello]
plt.title("Cakes selling in %")
plt.pie(y,labels=x,autopct="%2.2f%%",startangle=90,colors=z,explode=w,shadow=True)
plt.legend(loc="lower right")
plt.show()
```

```python
#bbox_to_anchor parameter specified the legent position so the tuple specifi the right adge 
import matplotlib.pyplot as plt
y=[300,300,400,950,800,750,600]
Toyname=["pikachoo","pegwin","taddey","rabit","dora","pepapig","OGY"]
col=["red","blue","yellow","green","orange","hotpink","lightgreen"]
r=[0,0,0,0.3,0,0,0]
title=("Mom's")
plt.pie(y,labels=Toyname,
        autopct="%2.2f%%",
        startangle=90,
        explode=r,
        colors=col,
        shadow=True)
plt.legend(loc="center left",title="Toys",bbox_to_anchor=(1,0,0.5,1))
plt.show()
       
        
```
